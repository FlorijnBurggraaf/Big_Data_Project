```python
import pandas as pd
```

```python
price_df = pd.read_csv('C:\\Users\\flori\\Downloads\\stocknet-dataset-master\\price\\raw\\MSFT.csv')
price_df.head()
```

```python
adj_close_list = price_df['Adj Close'].tolist()
target_list = [0]
for i in range(1,len(adj_close_list)):
    if adj_close_list[i] < adj_close_list[i-1]:
        target_list.append(0)
    else:
        target_list.append(1)
price_df['target'] = target_list
price_df.head()
```

```python
price_df = price_df[(price_df['Date'] >= '2014-01-01') & (price_df['Date'] <= '2016-03-31')] #Filter to the relevant time range
price_df.drop(columns=['Open','High','Low','Close','Adj Close','Volume'], inplace=True) #Drop irrelevant columns
price_df.head()
```

```python
price_df.to_csv('prediction_targets.csv', encoding= 'utf-8', index=None)
```
